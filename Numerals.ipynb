```python
import pandas as pd
from utilities import Codepoint
```

```python
df = pd.DataFrame(
    [
        (cp, cp.character, cp.script, cp.to_numeric(None), cp.to_digit(None), cp.to_decimal(None))
        for cp in Codepoint.iterate_all_codepoints()
    ], 
    index=[cp.value for cp in Codepoint.iterate_all_codepoints()],
    columns=[
    'Codepoint', 'Character', 'Script', 'Numeric', 'Digit', 'Decimal'
    ]
)


df
```
